```python
def icfun(x):
    tinit = 919
    tsurr = 500
    if x == 0:
       t = 500
    elif x > 0 and x <0.1:
       t = 500 + (919-500)/0.1*x
    elif x >= 0.1 and x < 14.9:
       t = 919
    elif x >= 14.9 and x < 15:
       t = 919 - (919-500)/0.1*(x-14.9)
    else:
       t = 500
    return t

x = np.linspace(0, 15, 100)
y = [icfun(i) for i in x]

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Temperature')
plt.title('Initial Condition')
plt.show()

    
```
